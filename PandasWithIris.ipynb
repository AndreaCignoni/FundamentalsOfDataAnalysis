#Investigation the Iris Dataset using Pandas

![Iris Flowers](https://s3.amazonaws.com/assets.datacamp.com/blog_assets/Machine+Learning+R/iris-machinelearning.png)

```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/illinois-cse/data-fa14/gh-pages/data/iris.csv")
```

```python
df
```

```python
# Selecting columns
df[['sepal_length','species']]
```

```python
# Selectiong rows.
df[2:6] # first number inclusive and last number excluded.
```

```python
df [['species','sepal_length']][2:6]
```

```python
# Pandas function "loc" uses labels and their numbers.
df.loc[2:6] # last number included.
```

```python
# This sintax returns all rows and the specified columns.
df.loc[:,'species']
```

```python
# This sintax will returns all rows and the two specified columns.
df.loc[:,['sepal_length','species']]
```

```python
# Slicing rows from 2 (inclusive) and 6 (inclusive) for the specified columns.
df.loc[2:6,['sepal_length','species']]
```

```python
# Slicing rows from 2 (inclusive) and 6 (inclusive) for all columns.
df.loc[2:6,:]
```

```python
# Creating a new data frame with the columns and rows specified.
newdf = df.loc[2:6,:]
```

```python
newdf
```

```python
df.iloc[2:6,:]
```

```python
# Panda function "iloc" uses positions and their rows number.
df.iloc[2] # accessing row 2
```

```python
# "iloc" function takes rows first and columns second as parameters.
df.iloc[2:6,2]
```

```python
# Taking every second row.
df.iloc[2:12:2,:]
```

```python
# Pandas function "at" takes labels as attributes
df.at[2,'species'] # extracting the species in the second row
```
